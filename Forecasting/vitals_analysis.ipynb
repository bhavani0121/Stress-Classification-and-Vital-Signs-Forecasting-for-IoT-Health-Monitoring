```python
import warnings
import itertools
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
plt.style.use('fivethirtyeight')
import pandas as pd
import statsmodels.api as sm
import matplotlib
from pandas import datetime
from matplotlib import pyplot
```

```python

def parser(t):
    return datetime.strptime(t,"\'[%H:%M:%S %d/%m/%Y]\'")


df = pd.read_csv("samples.csv",header=0,index_col=0,parse_dates=['Time and date'] , date_parser = parser,squeeze=True) 
df.head(5)
```

```python
plt.style.use('classic')
%matplotlib inline
plt.figure(figsize=(7,5))
df.plot()
plt.title('Pulse')
plt.grid()
plt.show()
```

```python
df.plot.kde()
plt.xlim([-10,150])
plt.title('Pulse - density plot before replacement')
df.std()
```

```python
cnt=0

for i in range(1,len(df)):
  if df[i]==0:
    cnt+=1
    df[i]=df[i-1]
df1=df
df1 = df1.resample('5S').mean()
#last 5mins data
print(cnt) #missing values
print(len(df))
```

```python
plt.style.use('classic')
%matplotlib inline
plt.figure(figsize=(7,5))
df.plot()
plt.title('Pulse')
plt.grid()
plt.show()
```

```python
df.plot.kde()
plt.xlim([-10,150])
plt.title('Pulse - density plot after replacement')
df.std()
```
